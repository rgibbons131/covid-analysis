```python
import numpy as np
import pandas as pd
import os
import urllib
```

```python
#Setting paths for the data
url = "https://covid19.who.int/WHO-COVID-19-global-data.csv"
file_path = os.path.join("data","covid")
```

```python
#This is retrieving the data and putting it in the folders we prepared
os.makedirs(file_path, exist_ok=True)
csv_path = os.path.join(file_path,"WHO-COVID-19-global-data.csv")
urllib.request.urlretrieve(url,csv_path)
```

```python
#This reads the data into a variable, called a dataframe
df= pd.read_csv(csv_path)
```

```python
#This just displays the data frame
df
```

```python
#This shows how many entrys there are
df_index = df.index
df_index
```

```python
#This shows the names of the columns in the data set
df_cols = df.columns
df_cols
```

```python
#This would take any white space out of the columns, allowing us to use beter methods
df.columns = [col.strip() for col in df.columns]
df.columns
```

```python
#We can set paramaters for the info we want to display, this displays all of the US data
df[df.Country == 'United States of America']
```

```python
#Get the info for just the US
us = df[df.Country_code == "US"]
us
```

```python
#Which day had the most new cases in the US
#We can see from the data above at time of writing the code, January of 2022 had the most new cases of Covid 19
us.loc[us.New_cases.idxmax(),["Date_reported","New_cases","Cumulative_cases"]]
```
